[Свертка списка](https://ru.wikipedia.org/wiki/%D0%A1%D0%B2%D1%91%D1%80%D1%82%D0%BA%D0%B0_%D1%81%D0%BF%D0%B8%D1%81%D0%BA%D0%B0) -  это обобщенная операция над списком, c помощью которой можно преобразовать список в единое значение. Например, рассмотрим реализации свертки слева и свертки справа (левоассоциативную свертку и правоассоциативную свертку):

```python
def foldl(f, x0, lst):
    if not lst:
        return x0
    return foldl(f, f(x0, lst[0]), lst[1:])

def foldr(f, x0, lst):
    if not lst:
        return x0
    return f(lst[0], foldr(f, x0, lst[1:]))
```

```python
#сумма
foldl(lambda x, y: x + y, 0, [1, 2, 3])
```

```python
#конкатенация
foldl(lambda x, y: '{}{}'.format(x, y), '', [1, 2, 3])
```

# Задание 2

нужно написать функцию, которая принимает две строки и проверяет, входит ли хотя бы одна перестановка второй строки в первую. Например:

> a = 'abcrotm'
> 
> b = 'tro'

функция def check_inv(a, b) вернет True, так как 'rot' содержится в 'abcrotm'. Нужно подумать как можно более оптимальный алгоритм и оценить его сложность. 

```python
def check_inv(a, b):
    a_dict = {a[i]: i for i in range(len(a))}
    max_index = a_dict[max(b, key = lambda x: a_dict[x])]
    min_index = a_dict[min(b, key = lambda x: a_dict[x])]
    if max_index - min_index == len(b) - 1:
        return True
    return False
check_inv('abcrotm', 'tro')
```

# Задание 4

Реализовать простейший калькулятор математических выражений:
- только целые числа
- **+**, **\-**, **\***, **\**
- скобки

**Можно использовать регулярные выражения**

```python
nums = '1234567890'
def get_num(expr):
    if expr[0] not in nums: return -1
    for i in range(len(expr)):
        if expr[i] not in nums:
            return i
    return 1

def calc(expr):
    col = []
    tex = []
    i = 0
    expr += ' '
    while(i < len(expr)):
        if(expr[i] == ' '):
            i += 1
            continue
        num = get_num(expr[i:-1])
        if num != -1:
            col.append(expr[i:i + num])
            i += num
        else: 
            if expr[i] == '+' or expr[i] == '-':
                if len(tex) == 0:
                    tex.append(expr[i])
                    i += 1
                    continue
                #1
                if tex[-1] == '(':
                    tex.append(expr[i])
                    i += 1
                    continue
                #2
                else:
                    col.append(tex[-1])
                    tex.pop()

            if expr[i] == '*' or expr[i] == '/':
                if len(tex) == 0:
                    tex.append(expr[i])
                    i += 1
                    continue
                #2
                if tex[-1] == '*' or tex[-1] == '/':
                    col.append(tex[-1])
                    tex.pop()
                    continue
                #1
                else:
                    tex.append(expr[i])
                    i += 1
                    continue
                    
            if expr[i] == '(':
                tex.append(expr[i])
                i += 1
                continue
            if expr[i] == ')':
                #3
                if tex[-1] == '(':
                    tex.pop()
                    i+=1
                    continue
                #2
                else:
                    col.append(tex[-1])
                    tex.pop()
    col.append(tex[-1])
    steck = []
    for i in col:
        if i[0] in nums:
            steck.append(int(i))
        else:
            if i == '+':
                steck[-2] += steck[-1]
                steck.pop()
            if i == '-':
                steck[-2] -= steck[-1]
                steck.pop()
            if i == '*':
                steck[-2] *= steck[-1]
                steck.pop()
            if i == '/':
                steck[-2] /= steck[-1]
                steck.pop()
    return steck[-1]
    

calc('2 * (15 - 3 * 4) - 2') == 4
#calc('(8+2*5)/(1+3*2-4)') 
```
